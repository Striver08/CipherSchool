```python
#To take input and check if the budget entered is float or not
while True:  
    try: 
        budget = float(input("Enter your budget : ")) 
        tempBudget = budget  #Storing budget in tempBudget so as to do calculations
    except ValueError: 
        print("The Entered Value should be Float") 
        continue
    else: 
        break
   
#grocery is the dict to store product details.

grocery ={"product":[],
    "quantity":[],
    "price":[]
   } 
  
groceryList = list(grocery.values())  # converted to list so that we can update if the product is present in the dict. 
  
prod = groceryList[0] 
qty = groceryList[1] 
pri = groceryList[2] 
  
# This loop terminates when user select 2.EXIT option when asked 
# in try it will ask user for an option as an integer (1 or 2)  
# if correct then proceed else continue asking options 


while True: 
    try: 
        choice = int(input("1.ADD \n 2.EXIT \n Enter your choice: ")) 
    
    except ValueError: 
        print("Choose only 1 or 2") 
        continue
    
    else: 
        
        if choice == 1 and tempBudget>0:   #If choice is 1, then using tempBudget to check if it is greater than zero or not     
               
            #Taking the Product details
            choicePro = input("Enter product name : ")
            choiceQty = input("Enter quantity : ") 
            choicePrice = float(input("Enter price of the product : "))   
  
            if choicePrice>tempBudget: # checking if price is less than budget and throwing a notification 
                print("You are out of your budget")  
                continue
  
            else: 

                if choicePro in prod:  # checking whether the product is already present in the list or not.  
                    
                    ind = prod.index(choicePro)   

                    qty.remove(qty[ind])  #Removing its details and adding the price in the tempBudget 
                    tempBudget = tempBudget + pri[ind]
                    pri.remove(pri[ind])   
  
                    qty.insert(ind, choiceQty)    #Inserting the new Product details.
                    pri.insert(ind, choicePrice)    
  
                    tempBudget = budget-sum(pri) #Adding all the prices and subtracting it from the Budget and storing in tempBudget
  
                    print("\n Amount left", tempBudget) 
                else: 
                    #Inserting the new Product details
                    prod.append(choicePro)   
                    qty.append(choiceQty)     
                    pri.append(choicePrice)     
  
                    tempBudget  = budget-sum(pri) #Adding all the prices and subtracting it from the Budget and storing in tempBudget
  
                    print("\n Amount left", tempBudget) 
  
        elif tempBudget<= 0:  # if budget is less than or equal to 0
            print("\n NO BUDGET")  
        else: 
            break 
        if choice ==2:
            break
  

print("\n Amount left : Rs.", tempBudget)  

#If some budget and some product price is equal to the budget--> Suggestion will be shown so that the user can buy it.
    
if tempBudget in pri:  
    print("\nAmount left can buy you a", prod[pri.index(tempBudget)])   
  
print("\n\n GROCERY LIST") 
  
print("Product Name", "Quantity", "Price") 
for i in range(len(prod)):  
    print(prod[i],"      ", qty[i],"   ", pri[i]) 
```
